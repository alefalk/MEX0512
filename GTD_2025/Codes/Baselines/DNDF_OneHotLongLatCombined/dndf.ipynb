```python
import sys
from itertools import product
from train import main  # Import your main function
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define search space
n_tree_values = [80, 100]            # example values
tree_depth_values = [10, 20]          # example values
batch_size_values = [1000]    # example values

# Results list
results = []

# Loop over all combinations
for n_tree, tree_depth, batch_size in product(n_tree_values, tree_depth_values, batch_size_values):
    print(f"Running with n_tree={n_tree}, tree_depth={tree_depth}, batch_size={batch_size}")

    # Set up sys.argv
    sys.argv = [
        'train.py',
        '-dataset', 'gtd478',
        '-n_class', '30',
        '-gpuid', '0',
        '-n_tree', str(n_tree),
        '-tree_depth', str(tree_depth),
        '-batch_size', str(batch_size),
        '-epochs', '100'
    ]

    # Train and evaluate
    y_true, y_pred = main()

    # Calculate metrics
    acc = accuracy_score(y_true, y_pred)
    prec_macro = precision_score(y_true, y_pred, average='macro')
    recall_macro = recall_score(y_true, y_pred, average='macro')
    f1_macro = f1_score(y_true, y_pred, average='macro')
    f1_micro = f1_score(y_true, y_pred, average='micro')

    # Save results
    results.append({
        'n_tree': n_tree,
        'tree_depth': tree_depth,
        'batch_size': batch_size,
        'accuracy': acc,
        'precision_macro': prec_macro,
        'recall_macro': recall_macro,
        'f1_macro': f1_macro,
        'f1_micro': f1_micro
    })

# After grid search: show best configuration
import pandas as pd

df_results = pd.DataFrame(results)
display(df_results)

# Find best by f1_macro for example
best_config = df_results.loc[df_results['f1_macro'].idxmax()]
print("Best configuration:")
print(best_config)
```
